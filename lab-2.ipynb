```python
"""
Задание 1.1.1
Считать три числа, найти минимальное и вывести.
"""
a = float(input("Введите первое число: "))
b = float(input("Введите второе число: "))
c = float(input("Введите третье число: "))

min_num = a
if b < min_num:
    min_num = b
if c < min_num:
    min_num = c

print("Минимальное число:", min_num)
```

```python
"""
Задание 1.1.2
Считать три числа, вывести те, которые попадают в интервал [1, 50].
"""
a = float(input("Введите первое число: "))
b = float(input("Введите второе число: "))
c = float(input("Введите третье число: "))

print("Числа в интервале [1, 50]:")
if 1 <= a <= 50:
    print(a)
if 1 <= b <= 50:
    print(b)
if 1 <= c <= 50:
    print(c)
```

```python
"""
Задание 1.1.3
Считать вещественное число m. Вывести последовательность 1*m, 2*m, ..., 10*m.
Использовать цикл for.
"""
m = float(input("Введите число m: "))

for i in range(1, 11):
    print(f"{i} * {m} = {i * m}")
```

```python
"""
Задание 1.1.4
Считать последовательность целых чисел (пока не будет введён 0).
Найти сумму и количество чисел, используя цикл while.
"""
print("Введите целые числа (0 для завершения):")
total = 0
count = 0

while True:
    num = int(input())
    if num == 0:
        break
    total += num
    count += 1

print("Сумма чисел:", total)
print("Количество чисел:", count)
```

```python
"""
Задание 2.8 (вариант 8)
Считать строку. Вывести количество слов в строке.
Без использования split().
"""
text = input("Введите строку: ")

word_count = 0
in_word = False

for i in text:
    if i != ' ' and not in_word:
        word_count += 1
        in_word = True
    elif i == ' ':
        in_word = False

print("Количество слов:", word_count)
```

```python
"""
Задание 3.8
Считать одномерный массив из N целочисленных элементов.
Вывести сумму, произведение.
Заменить нули на среднее арифметическое.
"""
vhod = input("Введите числа через пробел: ")

array = []
temp = ""

for i in vhod:
    if i != ' ':
        temp = temp + i
    else:
        if temp != "":  # если есть что добавить
            array.append(int(temp))
            temp = ""  # очищаем временную переменную

# Добавляем последнее число (после цикла)
if temp != "":
    array.append(int(temp))

print(f"Исходный массив: {array}")

# 1. Сумма
summa = 0
for i in array:
    summa = summa + i
print(f"Сумма: {summa}")

# 2. Произведение
proizvedenie = 1
for i in array:
    proizvedenie = proizvedenie * i
print(f"Произведение: {proizvedenie}")

# 3. Среднее арифметическое
srednee = summa / len(array)
print(f"Среднее: {srednee}")

# 4. Заменяем нули на среднее
noviy_massiv = []
for i in array:
    if i == 0:
        noviy_massiv.append(srednee)
    else:
        noviy_massiv.append(i)

print(f"Новый массив: {noviy_massiv}")
```
